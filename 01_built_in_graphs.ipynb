### Scale free graph

无标度网络（Scale-free Graph）是一种特殊类型的图结构，其度分布遵循幂律分布（Power Law Distribution）。在无标度网络中，只有少数几个节点具有非常高的度数，而大多数节点的度数相对较低。

具体来说，无标度网络具有以下特征：

* 幂律度分布：无标度网络的度分布服从幂律分布，即节点的度数与其度的频率成反比。这意味着只有少数几个节点具有非常高的度数，而大多数节点的度数相对较低。这种分布特征使得无标度网络在结构上具有高度不均衡性。

* 偏好连接：无标度网络的形成受到节点的偏好连接机制的影响。具有高度连接的节点更有可能吸引新的连接，形成一个富者恒富的现象。这种偏好连接机制使得网络中的一些节点逐渐积累了更多的连接，而其他节点则相对较少。

* 小世界性：无标度网络通常具有小世界性质，即节点之间的平均路径长度相对较短。尽管存在一些高度连接的节点，整个网络仍然具有较短的平均路径长度，使得节点之间的信息传播更加高效。

无标度网络在多个领域中都具有重要应用。例如，社交网络、互联网、生物网络等都被观察到具有无标度特性。无标度网络的研究有助于我们理解网络结构的形成机制、信息传播的路径以及系统的鲁棒性和弹性等方面。

总而言之，无标度网络是一种具有幂律度分布和偏好连接特性的图结构，其度数分布高度不均衡，只有少数节点具有非常高的度数。这种网络在许多领域中都有实际应用，并且对于理解复杂网络的形成和功能具有重要意义。

### Social Network

#### Karate Club Dataset

```python
# Karate Club Dataset
G = nx.karate_club_graph()
nx.draw_networkx(G)
```

空手道俱乐部数据集（Karate Club Dataset）是一份经典的社交网络数据集，用于研究社交网络和社群发现算法。该数据集描述了一个20世纪70年代早期的美国大学空手道俱乐部的成员之间的关系网络。

该数据集由Zachary Karate Club问题而得名，是由社会网络分析领域的研究者Wayne W. Zachary于1977年收集的。数据集中包含34个节点（空手道俱乐部的成员）和78条边（成员之间的相互关系）。每个节点代表俱乐部的成员，而边表示成员之间的社交关系或互动。

该数据集通常用于测试社群发现算法和社交网络分析方法。最为著名的应用是在研究社区划分问题上，即将俱乐部成员划分为两个社区，这在原始数据中反映了一个真实的分裂事件。根据数据集中的成员信息和社交关系，可以通过社区发现算法将成员分为两个相对独立的社区。

空手道俱乐部数据集对于研究社交网络的结构、社群发现算法的性能评估以及网络中社交互动的动态过程都非常有用。它成为了社会网络分析和复杂网络研究领域中的一个经典案例，为研究人员提供了一个标准的基准数据集。

NodeView 是网络分析中常见的数据类型之一，它表示了网络图中的一个节点或一组节点的视图或子图。

在网络图中，节点表示图中的个体或实体，而边表示节点之间的关系或连接。NodeView 是一个以节点为中心的数据结构，它提供了对单个节点或一组节点的访问和操作方法。

NodeView 可以用于从整个网络图中提取子图或局部图。通过选择特定的节点或基于某些条件过滤节点，可以创建一个 NodeView，其中包含了与选定节点相关联的节点和边。这使得我们可以对子图进行更深入的分析，而不必考虑整个网络。

NodeView 还可以提供有关节点的属性和邻居节点的信息。通过 NodeView，我们可以查看节点的属性、度数、邻居节点以及与节点相关的其他图结构信息。这些信息对于网络分析、社交网络分析、图算法等任务非常有用。

NodeView 可以是不可变的或可变的，取决于具体的网络分析库或工具。不可变的 NodeView 通常提供只读的属性和方法，而可变的 NodeView 则允许对节点和邻居进行修改。

总而言之，NodeView 是一个表示网络图中节点或一组节点的视图或子图的数据类型。它提供了对节点的属性、邻居节点和与节点相关的图结构信息的访问和操作方法，用于网络分析和图算法的任务。具体实现和功能会根据所使用的网络分析库或工具而有所不同。
