## NetworkX built-in graphs creation

### Fully-connected undirected graph

```python
G = nx.complete_graph(7)
nx.draw_networkx(G)
```

```python
# number of edges
G.size()
```

### Fully-connected directed graph

```python
G = nx.complete_graph(7, nx.DiGraph)
nx.draw_networkx(G)
```

```python
G.size()
```

```python
G.is_directed()
```

### Cycle graph

```python
G = nx.cycle_graph(6)
nx.draw_networkx(G)
```

### Ladder graph

```python
G = nx.ladder_graph(5)
nx.draw_networkx(G)
```

### Path graph

```python
G = nx.path_graph(20)
nx.draw_networkx(G)
```

### Star graph

```python
G = nx.star_graph(7)
nx.draw_networkx(G)
```

```python
G = nx.wheel_graph(8)
nx.draw_networkx(G)
```

### Binomial tree

```python
G = nx.binomial_tree(5)
nx.draw_networkx(G)
```

```python
G = nx.grid_2d_graph(3,5)
nx.draw_networkx(G)
```

#### Les miserables graph

```python
G = nx.les_miserables_graph()
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(G, seed=10)
nx.draw_networkx(G, pos)
```
